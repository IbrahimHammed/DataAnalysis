```python
# Method-3 : Using query
df.query('`Cust ID` == 1029312')
```

```python
df.query('`Cust ID` == 1029312')[['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']]
```

```python
df.query("Age == 44")[['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']]
```

```python
df.query('Age >= 20 and Age <= 30')[['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']]
```

```python
df.query("Status == 'Delivered' and Gender == 'Women' and Age < 20 and Amount < 280")[['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']]
```

```python
x = ['S', 'M', 'L']
df.query('Size in @x')[['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']]
```

```python
# Method 4: Using list
df.Size.unique()
```

```python
filterParameter = ((df['Month']=='Dec') & (df['Age'] < 20))
xColumns = ['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']

df[filterParameter][xColumns].head()
```

```python
xFilter = ['XL', 'M', 'Free']
xColumns = ['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']

df[df["Size"].isin(xFilter)][xColumns]
```

```python
# Method 5: Using filter conditions
xFilter = df["Age"] == 20
xColumns = ['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']
df[xFilter][xColumns]
```

```python
xFilter = (df["Age"] == 20) & (df["Size"] == 'XL')
xColumns = ['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']
df[xFilter][xColumns]
```

```python
xFilter = (df["Age"] == 20) & (df["Size"] == 'XL') & ((df["Amount"] > 400) & (df["Amount"] <= 500))
xColumns = ['Order ID', 'Date', 'Age', 'Age Group', 'Gender', 'Status', 'SKU', 'Size', 'Qty', 'Amount']
df[xFilter][xColumns]
```
