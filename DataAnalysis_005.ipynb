**Data cleansing techniques are demonstrated in this notebook**

```python
import pandas as pd
import numpy as np
```

```python
Student = ['Lola','Lara','Oluchi','Fatimah','Labake','Onyinye','Lateefah','Jennifer','Lisa','Kristin','Laura','Zainab','Ebube','Omotola','Fernanda','Isabella','Anett','Linda','Kashifah','Fatimah']
Gender = ['F','F',np.nan,'F','F',np.nan,np.nan,'F','F','F',np.nan,'F','F','F','F',np.nan,'F','F',np.nan,'F']
Age = [np.nan, 18, np.nan, np.nan, 18, 19, 18, 19, 18, np.nan, np.nan, 19, 18, 17, 18, 17, 19, 20, 19, np.nan]
Biology = [30, 27, 18, np.nan, np.nan, 24, 21, 26, np.nan, 19, 27, 25, np.nan, 20, 21, 19, 28, 22, 19, np.nan]
Chemistry = [24, 16, 19, np.nan, 21, np.nan, 15, 17, 20, np.nan, 18, 19, 21, 17, np.nan, 16, 18, 20, 19, np.nan]
Physics = [23, 21, np.nan, np.nan, 20, np.nan, 26, 24, 19, 21, 27,np.nan, np.nan, 22, 28, 20, np.nan, 18, 24, np.nan]

data = {'Student':Student, 'Gender':Gender, 'Age':Age, 'Biology':Biology, 'Chemistry':Chemistry, 'Physics':Physics}
```

```python
dfCA = pd.DataFrame(data)
dfCA
```

```python
dfCA[dfCA.duplicated()]
```

```python
dfCA.drop_duplicates(inplace=True)
dfCA
```

**Handling missing enties**

```python
# Method-1: Delete records affected
dfNew = dfCA.dropna()
dfNew.shape
```

```python
dfNew
```

```python
# Method-2: Replace with a known text value [for text columns]
# Option-1
dfNew = dfCA.fillna(0, inplace=False)
dfNew
```

```python
# Method-2: Replace with a known text value [for text columns]
# Option-2
dfNew = dfCA['Gender'].fillna('F', inplace=False)
dfNew
```

```python
# Method-5: Replace with mean of the specific column [for numeric columns]
dfCA.head()
```

```python
arbitraryNumber = dfCA['Age'].mean()
arbitraryNumber
```

```python
dfCA['Age'].fillna(arbitraryNumber, inplace=True)

# Converting Age column from float to int
dfCA['Age'] = dfCA['Age'].astype(int)
dfCA
```

```python
# Method-6: Backfill
dfCA.fillna(method='bfill', inplace=True)
dfCA
```

```python
# Focusing on a specific column
dfCA['Biology'].fillna(method='bfill', inplace=True)
```

```python
# Method-7: Forwardfill
dfCA.fillna(method='ffill', inplace=True)
dfCA
```

```python
# Method-8: Interpolate
dfCA.interpolate(method='linear', inplace=True)
dfCA
```
