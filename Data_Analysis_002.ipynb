```python
# Selects desired columns from the long list of 18
desiredColumns = ['Order ID', 'Cust ID', 'Gender', 'Age', 'SKU', 'Size', 'Qty', 'Amount']

# Apply the filter to the dfRetail to produce another dataframe
dfSummary = dfRetail[desiredColumns]
```

```python
# Preview the summary dataframe
dfSummary.head()
```

```python
# Generate a new column based on 2 existing ones
dfSummary["UnitPrice"] = dfSummary["Amount"] / dfSummary["Qty"]
```

```python
# Filter records further to produce 3 numeric columns
desiredColumns = ['Age','Qty', 'Amount']
dfCorrelations = dfRetail[desiredColumns]
dfCorrelations.head()
```
