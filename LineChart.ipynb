```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SalesMonth = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June']
SalesValue = [18, 28, 21, 29, 24, 29]

Results = {"Month": SalesMonth, "Sales": SalesValue}
df = pd.DataFrame(Results)
df
```

```python
# Line, Bar, Horizontal Bar, Pie, Donut, Scatter, Stacked Bar
# Audience Analysis
# Data nature
# Story line
```

```python
fig, ax = plt.subplots()

# Anytime you use ax.plot you'll end up with a line chart
ax.plot(
    # Indicate the field thta goes into your x-axis
    df['Month'],

    # Indicate the field thta goes into your y-axis
    df['Sales'],

    # Change Line type
    ls="-.",

    marker="o",
    markersize=10,
    markerfacecolor="red",
    markeredgecolor="black",

    # Change the width of the line
    linewidth=2
    )

# This sets the lower and upper limits for y axis values
ax.set_ylim(0,50)

# Insert chart title
plt.title("Half-Year Sales Summary for 2024", fontsize="10", color="navy")

# Insert x-axis label
plt.xlabel("Sales Month", fontsize="x-small", color="#452DEA")

# Insert y-axis label
plt.ylabel("Sales (in million $)", fontsize="small", color="#452DEA") # Hex notation for color

# Increase the chart size (x, y) in inches
plt.figure(figsize=(4, 4))

plt.show()
```
