```python
def mapTime(x):
  if x == 'Morning':
    return '08:00:00'
  elif x == 'Afternoon':
    return '13:00:00'
  elif x == 'Night':
    return '22:00:00',
  elif x == 'Early morning':
    return '06:00:00',
  elif x == 'Overnight':
    return '02:00:00',
  elif x == 'Unknown':
    return '01:00:00',
  elif x == '****':
    return '01:00:00',
  elif x == 'evening':
    return '19:00:00',
  elif x == 'Late afternoon':
    return '15:00:00',
  elif x == 'Before dawn':
    return '04:00:00',
  elif x == 'Late night':
    return '23:00:00',
  elif x == 'Late evening':
    return '20:00:00',
  elif x == 'Evening':
    return '19:00:00',
  elif x == 'Late morning':
    return '11:00:00',
  elif x == 'overnight':
    return '02:00:00',
  elif x == ' ':
    return '01:00:00',
  else:
    return x
```
