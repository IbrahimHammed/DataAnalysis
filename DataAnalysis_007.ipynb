```python
# Form data frame
dfSalesData = pd.DataFrame({'Date': dateVals, 'Sales': salesVals})
dfSalesData.sample(10)
```

```python
# Total Sales
dfSalesData["Sales"].sum()
```

```python
# Average Sales
dfSalesData["Sales"].mean()
```

```python
# Highest Sales
dfSalesData["Sales"].max()
```

```python
# Minimum Sales
dfSalesData["Sales"].min()
```

```python
# Transaction count
dfSalesData["Sales"].count()
```

```python
# Preview the dataframe
dfSalesData.head()
```

```python
# This does cummulative sum of the sales figures
dfSalesData["CummulativeSales"] = dfSalesData["Sales"].cumsum()
dfSalesData.head(10)
```

```python
# This demonstrates how to do 2 month-moving average with Python
MA = 2
dfSalesData["Sales2MA"] = dfSalesData["Sales"].rolling(window=MA).mean()
dfSalesData.head(10)
```

```python
# This demonstrates how to do 3 month-moving average with Python
MA = 3
dfSalesData["Sales3MA"] = dfSalesData["Sales"].rolling(window=MA).mean()
dfSalesData.head(10)
```

```python
# This gives date diff along the column
dfSalesData["SalesDateDiff"] = dfSalesData["Date"].diff(periods=1)
dfSalesData.head()
```

```python
# Gives the first day for whichever date is processed
dfSalesData["MonthFirstDay"] = dfSalesData["Date"].values.astype('datetime64[M]')
dfSalesData.head()
```

```python
# Shows data types of different columns
dfSalesData.dtypes
```

```python
# Changes the data type of the sales column in preparation for next set of analysis
dfSalesData["Date"] = dfSalesData["Date"].dt.to_timestamp()
dfSalesData.head()
```

```python
# Deducts date x from date y
dfSalesData["Duration"] = dfSalesData['Date'] - dfSalesData['MonthFirstDay']
dfSalesData.head()
```

```python
# Deducts 4 days from sales date column
dfSalesData["DateMinus4"] = dfSalesData["Date"] - pd.Timedelta(days=4)
dfSalesData.head()
```

```python
# Deducts 4 days from sales date column
dfSalesData["DateMinus4B"] = dfSalesData["Date"] + pd.Timedelta(days=-4)
dfSalesData.head()
```

```python
# Extracts the day part of sales date
dfSalesData["SalesDay"] = dfSalesData["Date"].dt.day
dfSalesData.head()
```

```python
# Extracts the day name part of sales date
dfSalesData["SalesDayName"] = dfSalesData["Date"].dt.day_name()
dfSalesData.head()
```

```python
# Extracts the month number part of sales date
dfSalesData["SalesDayMonth"] = dfSalesData["Date"].dt.month
dfSalesData.head()
```

```python
# Extracts the month name part of sales date
dfSalesData["SalesDayMonthName"] = dfSalesData["Date"].dt.month_name()
dfSalesData.head()
```

```python
# Extracts the quarter number of sales date
dfSalesData["SalesQuarter"] = dfSalesData["Date"].dt.quarter
dfSalesData.head()
```

```python
# Extracts the quarter part of sales date in the form of Q1
dfSalesData["SalesQuarterName"] = "Q" + dfSalesData["SalesQuarter"].astype(str)
dfSalesData.head()
```

```python
# Extracts the year part of sales date
dfSalesData["SalesYear"] = dfSalesData["Date"].dt.year
dfSalesData.head()
```
